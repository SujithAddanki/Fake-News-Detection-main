```python
import tensorflow as tf

# Load the trained model
model = tf.keras.models.load_model("best_model.keras")
```

```python
import sounddevice as sd
import numpy as np
import librosa
import tensorflow as tf
while 1:
    duration = 3 # Record for 3 seconds
    sample_rate = 22050  # Sampling rate
    
    def record_audio(duration=3, sample_rate=22050):
        print("Recording...")
        audio = sd.rec(int(duration * sample_rate), samplerate=sample_rate, channels=1, dtype='float32')
        sd.wait()  # Wait for recording to finish
        print("Recording complete.")
        return np.squeeze(audio)  # Remove unnecessary dimensions
    
    def predict_emotion(audio, sample_rate=22050):
        
        # Extract MFCC features
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
        mfccs = np.mean(mfccs, axis=1)  # Take mean across time
        mfccs = np.expand_dims(mfccs, axis=0)  # Add batch dimension
        
        # Make prediction
        prediction = model.predict(mfccs)
        predicted_class = np.argmax(prediction)  # Get class with highest probability
        
        # Map class index to emotion (Modify based on your dataset)
        emotions = ["Neutral", "Happy", "Sad", "Angry", "Fear", "Surprise"]
        print(f"Predicted Emotion: {emotions[predicted_class]}")
    
    # Run real-time emotion detection
    audio_data = record_audio(duration, sample_rate)
    predict_emotion(audio_data, sample_rate)
```
